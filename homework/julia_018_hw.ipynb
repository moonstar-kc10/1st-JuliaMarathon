## Day 018 作業1：The Split-Apply-Combine Strategy

同 Day 017，請載入 2020/4/2 COVID-19 Daily Report 資料集，計算每個國家/地區的累計確診人數，並列出所有資料筆數。

```julia
using DataFrames, CSV
```

```julia
df = CSV.read("04-02-2020.csv")
```

計算計算每個國家/地區的累計確診人數

【提示】可使用 groupby 相關的函式

```julia
gd_country = groupby(df, :Country_Region)
```

```julia
x = keys(gd_country)[1]
print(x)
```

```julia
for i in 1:181
    x = keys(gd_country)[i]
    total = sum(gd_country[i][:,:Confirmed])
    println("$(x),確診人數,$(total)")
end
```
